For two strings s and t, we say "t divides s" if and only if s = t + t + t + ... + t + t (i.e., t is concatenated with itself one or more times).

Given two strings str1 and str2, return the largest string x such that x divides both str1 and str2.

 

Example 1:

Input: str1 = "ABCABC", str2 = "ABC"

Output: "ABC"

Example 2:

Input: str1 = "ABABAB", str2 = "ABAB"

Output: "AB"

Example 3:

Input: str1 = "LEET", str2 = "CODE"

Output: ""

Example 4:

Input: str1 = "AAAAAB", str2 = "AAA"

Output: ""​​​​​​​

 

Constraints:

    1 <= str1.length, str2.length <= 1000
    str1 and str2 consist of English uppercase letters.



### The most efficient Solution

```python
class Solution:
    def gcdOfStrings(self, str1: str, str2: str) -> str:
        if str1 + str2 == str2+str1:
            x = gcd(len(str1),len(str2))
            return str1[:x]
        else:
            return ""
```

The intuition here is:
Example: str1 = ABCABC and str2 = ABC

both are equal, so we will just get the gcd of their lenghts and then return the string of that size

### Intuitive Solution

```python
class Solution:
    def gcdOfStrings(self, str1: str, str2: str) -> str:
        z = 1
        ls1 = len(str1)
        ls2 = len(str2)
        res = ""
        while z <= ls2:
            substr = str2[:z]
            if ls2 % z == 0:
                sub1 = ls1//z
                sub2 = ls2//z
  
                if (((substr * sub1) == str1) and ((substr * sub2) == str2)):
                    if len(substr) > len(res):
                        res = substr
           
            z += 1

        return res
```

Generate and Verify: It treats every possible prefix of the second string as a potential "building block," checking if repeating that block can perfectly reconstruct both original strings.

Maximize: By checking every length from 1 up to the end, it ensures that while it might find smaller common divisors first, the final result stored in res is guaranteed to be the longest one.
